```python
# Day-9
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import remove_stopwords, preprocess_string, strip_punctuation
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')


stop_words = set(stopwords.words('english'))


stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()


file_path = "spam.csv"
with open(file_path, 'r') as file:
    text = file.read()


text_no_stopwords = remove_stopwords(text)

tokens = simple_preprocess(text_no_stopwords, deacc=True)


stemmed_tokens = [stemmer.stem(token) for token in tokens]


lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]


print("Original Text:")
print(text)

print("\nText After Removing Stopwords:")
print(text_no_stopwords)

print("\nTokens:")
print(tokens)

print("\nStemmed Tokens:")
print(stemmed_tokens)

print("\nLemmatized Tokens:")
print(lemmatized_tokens)
```
